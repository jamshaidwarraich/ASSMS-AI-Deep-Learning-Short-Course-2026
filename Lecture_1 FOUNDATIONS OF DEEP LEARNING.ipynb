<div style="font-family: 'Times New Roman', Times, serif; background-color: #fdfaf5; border-left: 8px solid #800000; border-radius: 8px; padding: 20px 30px;"><h2 style="color: #800000; font-size: 24px; margin: 0 0 12px 0;">Variables in Python</h2><p style="font-size: 15px; color: #333; margin: 6px 0;">A variable is simply a name that stores a value. Python automatically figures out the data type based on what you assign.</p><ul style="font-size: 14px; color: #333; line-height: 1.8;"><li><strong style="color: #c0392b;">Integer:</strong> a whole number, such as Students</li><li><strong style="color: #2980b9;">Float:</strong> a decimal number, such as height</li><li><strong style="color: #27ae60;">String:</strong> text, written inside quotes, such as name</li></ul><p style="font-size: 14px; color: #555; margin-top: 10px;">The example below creates one variable of each type and prints them.</p></div>

```python
# Integer variable
age = 32

# Float variable
height = 6.1

# String variable
name = "Ali"

# Printing variables
print("Name:", name)
print("Age:", age)
print("Height:", height)
```

<div style="font-family: 'Times New Roman', Times, serif; background-color: #fdfaf5; border-left: 8px solid #800000; border-radius: 8px; padding: 20px 30px;"><h2 style="color: #800000; font-size: 24px; margin: 0 0 12px 0;">Checking Variable Types</h2><p style="font-size: 15px; color: #333; margin: 6px 0;">Python has a built in function called <code style="background-color: #f0e6e6; padding: 2px 6px; border-radius: 4px;">type()</code> that tells you what kind of data a variable holds.</p><p style="font-size: 14px; color: #555; margin-top: 10px;">The example below checks the type of each variable created earlier: name, height, and age.</p></div>

```python
print(type(name))
print(type(height))
print(type(age))
```

<div style="font-family: 'Times New Roman', Times, serif; background-color: #fdfaf5; border-left: 8px solid #800000; border-radius: 8px; padding: 20px 30px;"><h2 style="color: #800000; font-size: 24px; margin: 0 0 12px 0;">Taking User Input</h2><p style="font-size: 15px; color: #333; margin: 6px 0;">Python's built in function <code style="background-color: #f0e6e6; padding: 2px 6px; border-radius: 4px;">input()</code> allows a program to ask the user for information while it is running.</p><p style="font-size: 14px; color: #555; margin-top: 10px;">Whatever the user types is stored as text (a string) in a variable, which can then be used later in the program.</p></div>

```python
# Get user input
name = input("Enter your name: ")

# Display the input
print("Hello,", name)
```

```python
# Get user input
who_are_u = input("Enter your name: ")

# Display the input
print( who_are_u, "is a student")
```
