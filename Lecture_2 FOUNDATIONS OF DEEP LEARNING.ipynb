<div style="font-family: 'Times New Roman', Times, serif; background-color: #f5f0ff; border-left: 8px solid #6c3483; border-radius: 8px; padding: 25px 30px;"><h2 style="color: #2c3e50; font-weight: bold; font-size: 24px; margin: 0 0 12px 0;">numpy.eye</h2><p style="font-family: monospace; background-color: #e0d0f5; color: #2c3e50; padding: 8px 12px; border-radius: 6px; display: inline-block; font-size: 14px;">numpy.eye(N, M=None, k=0, dtype=&lt;class 'float'&gt;)</p><p style="font-size: 16px; color: #333; margin: 15px 0;">Returns a 2D array with ones on the diagonal and zeros elsewhere.</p><ul style="font-size: 16px; color: #333; line-height: 2;"><li><strong style="color: #6c3483;">N</strong> <span style="color: #888; font-size: 14px;">(int):</span> number of rows in the output</li><li><strong style="color: #2980b9;">M</strong> <span style="color: #888; font-size: 14px;">(int, optional):</span> number of columns in the output. If None, defaults to N</li><li><strong style="color: #c2185b;">k</strong> <span style="color: #888; font-size: 14px;">(int, optional):</span> index of the diagonal. 0 refers to the main diagonal, a positive value refers to an upper diagonal, and a negative value to a lower diagonal</li><li><strong style="color: #16a085;">dtype</strong> <span style="color: #888; font-size: 14px;">(data type, optional):</span> data type of the returned array</li></ul></div>

```python
# Create a 2x2 identity matrix
#eye(rows, columns, k=0, dtype)
# main diagonal
a=np.eye(2,2,0)
print("Diagonal matrix:\n",a)
# upper diagonal
a1=np.eye(4,4,2)
print("Upper diagonal matrix:\n",a1)
# lower diagonal
a2=np.eye(2,2,-1)
print("Lower Diagonal matrix:\n",a2)

d=2*np.eye(4,4,-1)+3*np.eye(4,4,1)
print("4x4 matrix with diagonal entries zeros:\n",d)
```

```python
d=2*np.eye(4,4,-1)+3*np.eye(4,4,1)+4*np.eye(4,4,0)
print("4x4 matrix with diagonal entries zeros:\n",d)
```

<h2 style="font-family: 'Times New Roman', Times, serif; color: #2c3e50; font-weight: bold; font-size: 26px;">Arithmetic Operations</h2>

```python
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a+b) # manually
print(np.add(a,b)) # by built in function
# Subtraction:
print(a-b) # manually
print(np.subtract(a,b)) # by built in function
# Multiplication
print(a*b) # manually
print(np.multiply(a,b)) # by built in function
# Division:
print(a/b) # manually
print(np.divide(a,b)) # by built in function
## Power:
print(a**2) # manually
print(np.power(a,2)) # by built in function
#remainder:
print(a%b) # manually
print(np.mod(a,b)) # by built in function
print(np.mod(b,a)) # by built in function

# Transpose:
print("Transpose:",a.T)
```

<div style="font-family: 'Times New Roman', Times, serif; background-color: #f5f0ff; border-left: 8px solid #6c3483; border-radius: 8px; padding: 20px 30px;"><h2 style="color: #2c3e50; font-weight: bold; font-size: 24px; margin: 0 0 10px 0;">Linear Algebra Functions</h2><p style="font-size: 16px; color: #333; margin: 0;">For matrix multiplication, the function <code style="background-color: #e0d0f5; color: #6c3483; padding: 2px 6px; border-radius: 4px;">np.matmul()</code> is used.</p></div>

```python
import numpy as np
## For Matrix Multiplication:
A=np.array([[[1,2], [3,4]]])
B=np.array([[[3,4], [3,7]]])

print(A.shape)
C=np.matmul(A,B)
print(C)
print("Inverse:",np.linalg.inv(A))   ## Inverse
print("Determinant:",np.linalg.det(A))   ## determinant
print("Eigen Values:",np.linalg.eig(A))   ## eigen values

```

<div style="font-family: 'Times New Roman', Times, serif; background-color: #f5f0ff; border-left: 8px solid #6c3483; border-radius: 8px; padding: 25px 30px;"><h2 style="color: #2c3e50; font-weight: bold; font-size: 26px; margin: 0 0 15px 0;">Array Indexing</h2><p style="font-size: 16px; color: #333; margin: 8px 0;">Indexing refers to the process of accessing individual elements or a subset of elements from a data structure, such as an array, list, or string, using a unique identifier known as an index.</p><p style="font-size: 16px; color: #333; margin: 8px 0;">The index refers to the position of an element within the structure, allowing it to be retrieved or modified directly.</p><p style="font-size: 16px; color: #333; margin: 8px 0;">In most programming contexts, indexing is zero based, meaning the first element is accessed using index 0.</p><p style="font-size: 16px; color: #333; margin: 8px 0;">In a one dimensional array, a single index is used, while in multi dimensional arrays, multiple indices are needed to specify the position of an element along each dimension.</p><p style="font-size: 17px; font-weight: bold; color: #6c3483; margin-top: 15px;">1D Array Indexing</p></div>

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])
print(arr[0])  # Output: 10
print(arr[3])  # Output: 40
print(arr[-1])
f=arr[0]=3# Output: 50 (negative indexing starting from the back of the array)
print(f)
```

```python
arr=np.array([[1,2,3,4],
              [5,62,1,2],
              [5,8,1,0]])
arr[1,2]=5 #if we want to
print(arr)
#print(arr[2,3])
```

<h2 style="font-family: 'Times New Roman', Times, serif; color: #2c3e50; font-weight: bold; font-size: 24px;">Array Indexing: Examples</h2>

```python
a=np.array([[1,2,3],[5,6,7],[4,0,1]])
a[1,1]
a[1,1]=9999
print(a)
print('==========================')

print('bbbb',a[:2])
print('==========================')

print(a[:2])
print('==========================')
print(a[1:])
# To print [3,9999,1]
print('==========================')

print('Required ele',[a[0,2],a[1,1],a[2,2]])
```

<div style="font-family: 'Times New Roman', Times, serif; background-color: #f5f0ff; border-left: 8px solid #6c3483; border-radius: 8px; padding: 20px 30px;"><h2 style="color: #2c3e50; font-weight: bold; font-size: 24px; margin: 0 0 10px 0;">Boolean Indexing</h2><p style="font-size: 16px; color: #333; margin: 6px 0;">Boolean array indexing lets you pick out arbitrary elements of an array.</p><p style="font-size: 16px; color: #333; margin: 6px 0;">Frequently, this type of indexing is used to select the elements of an array that satisfy some conditions. Here is an example:</p></div>

```python
a=np.array([1,6,8,0,2,5,7,8,3,4,9])
bool_index=(a > 4)
print(bool_index)
print(a[bool_index])
```

```python
myli=[0,3,5,7,2,1,8]
print(myli[:4])
print(myli[:-3])
print(myli[:1:-3])
```
