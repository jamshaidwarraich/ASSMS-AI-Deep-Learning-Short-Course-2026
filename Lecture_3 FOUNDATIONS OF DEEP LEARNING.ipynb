***Question*** : implements the following logic:

1. **Variable:**
   - `sunlight` ( `True` or `False`).

2. **Logic:**
   - If `sunlight` is `True`, print "It is day time".   

   - If `sunlight` is `False`, print "It is night".

The task is to write the code that evaluates the `sunlight` variable \
and produces the appropriate output based on whether it is daytime or night.


```python
# It is daytime
sunlight= True

if sunlight:
    print('It is day time')
if not sunlight:
    print('It is night')
```

```python
# It is daytime
sunlight= False

if sunlight:
    print('It is day time')
else:
    print('It is night')
```

**Question:**

Write a Python code that determines if the ASSMS is closed or open based on whether today is a holiday. 

1. **Variable:**
   - `holiday` (a boolean indicating if today is a holiday, `True` or `False`).

2. **Logic:**
   - If `holiday` is `True`, print "ASSMS is closed today".
   - If `holiday` is `False`, print "ASSMS is open today".

Your task is to write the code that evaluates the `holiday` variable and produces the appropriate message based on whether the school is closed or open.


```python
# Check if today is a holiday
holiday = False

if holiday:
    print('ASSMS is closed today')
else:
    print('ASSMS is open today')
```

Question:

        If neither electricity nor sunlight is True, print "we are not working".


```python
electricity= False
sunlight=True

if electricity and  sunlight:
        print('we are working')
        
elif electricity or sunlight:     
    
    print('we can work')
else:
    print('we are not working')
```
