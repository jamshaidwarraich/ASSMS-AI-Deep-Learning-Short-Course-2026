A line plot is one of the most commonly used graphs in Matplotlib. It connects data points with straight lines, making it easy to observe trends, patterns, and changes in data.

</p>

</div>

```python
## drawing a Simple line
x = [2, 4, 6, 8]     #Defines the values for the x-axis.
y = [1, 2, 3, 4]     #Defines the values for the y-axis.

plt.plot(x, y)      #Creates a line plot by connecting the data points.
plt.show()          #Displays the graph on the screen.
```

```python
#Adding Axis Labels

plt.xlabel("x-axis")         #Adds a label to the horizontal (x) axis.
plt.ylabel("y-axis")         #Adds a label to the vertical (y) axis.

plt.plot(x, y)
plt.show()
```

```python
#Adding a Title to figure
plt.title("Diagram 01")      # Add a title to the graph

plt.plot(x, y)
plt.show()
```

```python
#Customizing the Figure
plt.figure(figsize=(3,3))    # Set the size of the figure (Width, Height)

plt.plot(x, y)
plt.show()
```

```python
# Customizing Line Style and Color
plt.plot(x, y,":",color = "red")     # Draw a red dotted line
plt.show()
```

```python
#drawing multiple lines between multiple points
x = np.array([1,2,6,8])              # Define data for the first line
y = np.array([4,5,2,10])

x1 = np.array([1,5,7,6])             # Define data for the second line
y1 = np.array([3,3,7,10])

x2 = np.array([2,5,9,6])             # Define data for the third line
y2 = np.array([3,6,7,4])

plt.figure(figsize=(4,4))
plt.title("Diagram : 05")

# Draw the first line and assign a label
plt.plot(x, y, ":", color="r", label="Line 1")

# Draw the second line and assign a label
plt.plot(x1, y1,"--", color="g", label="Line 2")

# Draw the third line and assign a label
plt.plot(x2, y2, color="b", label="Line 3")

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```

```python
#Creating Subplots in Matplotlib

## Create the first subplot
x = np.array([1,2,6,8])
y = np.array([4,5,2,10])
plt.subplot(1,2,1)              #(no of rows,no of columns,position)
plt.plot(x,y,color="r")
plt.title("Plot 1")

## Create the second subplot
x = np.array([2,5,3,8])
y = np.array([4,7,2,17])
plt.subplot(1,2,2)
plt.plot(x,y,color="g")
plt.title("Plot 2")

# Add a title for the entire figure
plt.suptitle("Two plots in one figure")
plt.show()
```

```python
#Creating Multiple Subplots
# Create a figure with 2 rows and 2 columns (4 subplots)

#adjust size
plt.subplots(2, 2, figsize=(10, 6))  #(rows, columns)

#plot 1
x = np.array([1,2,6,8])
y = np.array([4,5,6,10])
plt.subplot(2,2,1)              #(rows,column,position)
plt.plot(x,y,color="r")
plt.title("Plot 1",color="b")

#plot 2
x = np.array([2,5,7,8])
y = np.array([4,7,8,9])
plt.subplot(2,2,2)
plt.plot(x,y,color="g")
plt.title("Plot 2",color="b")

#plot 3
x = np.array([4,6,10,16])
y = np.array([9,7,12,17])
plt.subplot(2,2,3)
plt.plot(x,y,color="b")
plt.title("Plot 3",color="b")

#plot 4
x = np.array([8,15,19,20])
y = np.array([4,8,10,11])
plt.subplot(2,2,4)
plt.plot(x,y,color="y")
plt.title("Plot 4",color="b")

#add suptitles
plt.suptitle("Four plots in one figure",fontsize=20)

# Adjust layout to prevent overlapping
plt.tight_layout(rect=[0, 0, 1, 0.95])     #(left, bottom, right, top)

plt.show()
```

Matplotlib allows us to display individual data points without connecting them with lines. This is useful when we want to visualize the exact position of each data point on the graph.

```python
#draw a coloured point with different edge color

x=np.array([1])
y=np.array([2])
plt.figure(figsize=(3,3))

plt.plot(x,y,'o',color="yellow",mec="black")      #Draw only the data points

plt.show()
```

```python
#star shape points

x = np.array([1,2,3,4,5])
y = np.array([4,2,6,3,8])
plt.figure(figsize=(4,4))

plt.plot(x, y,marker='*',linestyle='None',markersize=15,color="red")

plt.title("Star Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#adding markers at Points

x = np.array([1,2,3,4,5])
y = np.array([5,3,7,4,8])
plt.figure(figsize=(4,4))

plt.plot(x, y,marker='D',linestyle='-',markersize=10,markerfacecolor="deepskyblue",markeredgecolor="navy")

plt.title("Diamond Markers")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#Changing Marker Style and Colors

x=np.array([1,3,5,7])
y=np.array([2,4,6,9])
plt.figure(figsize=(3,3))

plt.plot(x,y,marker='s',color='red',linestyle=':',ms=10,mec='purple',mfc='pink')

plt.show()
```

A bar chart is used to compare values across different categories. In Matplotlib, bar charts are created using the <b style="color:#2F5D50;">plt.bar()</b> function, where the height of each bar represents its corresponding value.

</p>

</div>

```python
#draw simple bars chart

# Categories
x = np.array(["A", "B", "C", "D"])

# Values
y = np.array([10, 25, 15, 30])

# Create a bar chart
plt.figure(figsize=(4,4))
plt.bar(x, y)


plt.title("Simple Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Changing Bar Colors and width, and Adding Edge Colors

plt.bar(x, y,color="skyblue", width=0.6, edgecolor="black", linewidth=2)

plt.show()
```

```python
#Using Multiple Colors for Bars

x = np.array(["A", "B", "C", "D", "E"])
y = np.array([18, 25, 12, 30, 22])
plt.figure(figsize=(4,4))

# Draw a bar chart with different colors
plt.bar(x, y,color=["red", "olive", "pink", "orange", "plum"],edgecolor="black")


plt.title("Bar Chart with Multiple Colors")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
#Horizontal Bar Charts

x = np.array(["Apple", "Banana", "Orange", "Mango"])
y = np.array([25, 18, 30, 22])
plt.figure(figsize=(5,4))

# Draw a horizontal bar chart
plt.barh(x, y,color="plum",edgecolor="purple",height=0.5)

plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Fruits")
plt.show()
```

Matplotlib can be used to visualize mathematical functions by plotting their corresponding x and y values. This helps in understanding the behavior and shape of different functions.

```python
#Linear Function

# Create x values
x = np.linspace(-5,5,100)

# Calculate y values
y = x

plt.figure(figsize=(5,4))
plt.plot(x,y,color="blue")
plt.title("Linear Function : y = x")
plt.xlabel("x")
plt.ylabel("y")

# Display grid
plt.grid(True)
plt.show()
```

```python
#Quadratic Function

# Define the quadratic function
def f(x):
    return x**2

# Generate evenly spaced x-values
x = np.linspace(-10,10,10000)

# Calculate the corresponding y-values
y = f(x)

plt.figure(figsize=(4,4))
plt.plot(x,y,label=r'$f(x)=x^2$')
plt.title(r'Graph of $f(x)=x^2$',color="red",fontsize=15)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#Adding grid to Exponential Function 

# Define the exponential function
def f(x):
    return np.exp(x)

# Generate evenly spaced x-values
x = np.linspace(-2, 2, 200)

# Calculate the corresponding y-values
y = f(x)

plt.figure(figsize=(5,4))
plt.plot(x, y, color="green", linewidth=2, label=r"$f(x)=e^x$")
plt.title(r"Graph of $f(x)=e^x$", color="darkgreen", fontsize=15)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Display grid lines
plt.grid(True, linestyle="--", alpha=0.6, color='y')

plt.legend()
plt.show()
```

```python
#Plotting Multiple Functions on the Same Graph

#generate values between -5 to 5
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
#x = np.linspace(-5,5,10000)

# Compute sin(x) and cos(x)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the plot
plt.figure(figsize=(6,4))  
plt.plot(x, y1, label=r'$y = \sin(x)$', color='blue')  
plt.plot(x, y2, label=r'$y = \cos(x)$', color='red')

# Title and labels
plt.title(r'Graph of $\sin(x)$ and $\cos(x)$', fontsize=15, color="black",y=1.1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# Add a legend
plt.legend()

# Display the graph
plt.show()
```

<b style="color:#6D1F2A;">mpl_toolkits.mplot3d</b> is a Matplotlib toolkit that enables the creation of three-dimensional (3D) visualizations. It provides the <b style="color:#6D1F2A;">Axes3D</b> class, which allows users to plot 3D surfaces, wireframes, scatter plots, line plots, and other 3D graphs using the x-, y-, and z-axes.

</p>

```python
#Creating a 3D Scatter Plot

# Generate random x, y, and z coordinates
x=np.random.randn(500)
y=np.random.randn(500)
z=np.random.randn(500)

fig=plt.figure(figsize=(7,6))

# Add a 3D subplot
ax=fig.add_subplot(111,projection='3d')

# Draw the 3D scatter plot
ax.scatter(x,y,z,c=z,cmap="rainbow",s=50)

# Add a title
ax.set_title("3D Galaxy")

plt.show()
```

```python
#Creating a 3D Spiral Curve

# Generate parameter values
t=np.linspace(0,20,500)

# Calculate the x, y, and z coordinates
x=np.cos(t) 
y=np.sin(t)
z=t

fig=plt.figure(figsize=(7,6))
ax=fig.add_subplot(111,projection='3d')

# Draw the 3D spiral
ax.plot(x,y,z,color="red",linewidth=3)

ax.set_title("3D Spiral")

plt.show()
```

```python
#Creating a 3D Mountain Surface

# Generate evenly spaced x and y values
x = np.linspace(-4,4,150)
y = np.linspace(-4,4,150)

# Create a meshgrid for x and y coordinates
X,Y = np.meshgrid(x,y)

# Compute the height of the mountain
Z = np.exp(-(X**2+Y**2))

fig = plt.figure(figsize=(7,6))
ax = fig.add_subplot(111,projection="3d")

# Draw the mountain surface
ax.plot_surface(X,Y,Z,cmap="terrain")

ax.set_title("3D Mountain")

plt.show()
```

```python
#Creating a 3D Colorful Wave

# Generate evenly spaced x and y values
x = np.linspace(-5,5,150)
y = np.linspace(-5,5,150)

# Create a meshgrid for x and y coordinates
X,Y=np.meshgrid(x,y)

# Calculate the wave function
Z=np.sin(X)*np.cos(Y)

fig=plt.figure(figsize=(7,6))
ax=fig.add_subplot(111,projection="3d")

# Draw the colorful surface
ax.plot_surface(X,Y,Z,cmap="plasma")

ax.set_title("Colorful Wave")
plt.show()
```

```python
#Creating a 3D Ripple Surface

x=np.linspace(-6,6,200)
y=np.linspace(-6,6,200)

X,Y=np.meshgrid(x,y)

# Calculate the z values using the ripple function
Z=np.sin(np.sqrt(X**2+Y**2))

fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection='3d')

# Add a 3D subplot
surf=ax.plot_surface(X,Y,Z,cmap="turbo",edgecolor="none")

# Draw the 3D ripple surface
fig.colorbar(surf,shrink=0.6)

ax.set_title("3D Ripple Surface")
plt.show()
```

```python
#3D surface plot

#define a fn
def f(x,y):
    return x**2 + y**2

#generate x,y values
x = np.linspace(-5,5,10000)
y = np.linspace(-5,5,10000)

#create meshgrid for X and Y
X,Y=np.meshgrid(x,y)

#compute Z values using the fn
Z = f(X,Y)

#create a 3D plot
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

#plot the surface
ax.plot_surface(X,Y,Z)

#add label and title
ax.set_title('3D surface plot')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

#display the plot
plt.show()
```
