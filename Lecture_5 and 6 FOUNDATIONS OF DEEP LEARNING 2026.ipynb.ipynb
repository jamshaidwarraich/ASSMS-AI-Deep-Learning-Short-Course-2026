<div style="font-family:Segoe UI,system-ui,sans-serif;font-size:16px;color:#33475a;line-height:1.7">
Two keywords do all the work: <code style="background:#e6f0f6;color:#0a3d62;padding:2px 7px;border-radius:4px">class</code> defines the blueprint,
and <code style="background:#e6f0f6;color:#0a3d62;padding:2px 7px;border-radius:4px">def</code> gives it a behaviour called a <b>method</b>.
</div>

```python
# STEP 1 — define the blueprint
class Dog:
    def bark(self):        # bark = a method (function inside a class)
        print('woof')
```

```python
# STEP 2 — build an object, then make it bark
german_shephard = Dog()
german_shephard.bark()     # woof
```

```python
# STEP 3 — same blueprint makes another object
a_shephard = Dog()
a_shephard.bark()          # woof
```

  🔑 <code style="background:#e6f0f6;color:#0a3d62;padding:2px 8px;border-radius:5px">self</code> = "this object's own". It is the first input of every method.
</div>

<div style="background:#0a3d62;border-radius:10px;padding:14px 24px;font-family:Segoe UI,system-ui,sans-serif;margin-top:10px">
  <span style="color:#FFB000;font-size:30px;font-weight:900">02</span>
  <span style="color:#FFFFFF;font-size:27px;font-weight:800;margin-left:12px">__init__ — give the object a memory</span>
</div>

<div style="font-family:Segoe UI,system-ui,sans-serif;font-size:16px;color:#33475a;line-height:1.7">
<code style="background:#e6f0f6;color:#0a3d62;padding:2px 7px;border-radius:4px">__init__</code> is the <b>constructor</b> — it runs the moment an object is created and stores its starting data.
</div>

```python
# STEP 1 — a Dog that remembers its name and temperature
class Dog:
    def __init__(self, petname, temp):    # constructor
        self.name = petname               # attribute (data)
        self.temprature = temp

    def status(self):                     # getter — show the data
        print('petname is', self.name)
        print('pet temprature is', self.temprature)

    def settemprature(self, temp):        # setter — change the data
        self.temprature = temp
```

```python
# STEP 2 — one blueprint, three different objects
german_shephard = Dog('German shephard', 35)
lessi           = Dog('lessi', 39)
a_shephard      = Dog('a', 40)
```

```python
# STEP 3 — each object remembers its own data
german_shephard.status()
lessi.status()
a_shephard.status()
```

```python
# STEP 1 — define the Bag blueprint
class Bag:
    def __init__(self, color, price, weight):
        self.color  = color
        self.price  = price
        self.weight = weight

    def status(self):
        print('The bag is of', self.color)
        print('the bag costs', self.price)
        print('the bag has weight', self.weight)

    def setcolor(self, color):
        self.color = color
```

```python
# STEP 2 — build the object and show it
nike = Bag('blue color', 2000, 40)
nike.status()
```

```python
# STEP 3 — change the colour, show again
nike.setcolor('red')
nike.status()
```

```python
# STEP 1 — a Student blueprint with real fields
class Student:
    def __init__(self, name, registration_no, subjects, CGPA, grade):
        self.name = name
        self.registration_no = registration_no
        self.subjects = subjects
        self.CGPA  = CGPA
        self.grade = grade

    def status(self):
        print('My name is:', self.name)
        print('My registration no is:', self.registration_no)
        print('I have chosen the subjects:', self.subjects)
        print('My CGPA is:', self.CGPA)
        print('I got the grade:', self.grade)

    def setname(self, name):
        self.name = name
    def setregistration_no(self, registration_no):
        self.registration_no = registration_no
```

```python
# STEP 2 — build a real student object
student = Student('Momina Arshad', 10,
                  ['Functional Analysis', 'Advanced Differential Equations',
                   'Partial Differential Equations', 'Advanced Linear Algebra'],
                  3.57, 'B+')
student.status()
```

```python
# STEP 3 — update name and registration no
student.setname('Amna')
student.setregistration_no(20)
student.status()
```

<div style="font-family:Segoe UI,system-ui,sans-serif;font-size:16px;color:#33475a;line-height:1.7">
The real power of classes: make <b>many objects</b>, store them in a <b>list</b>, and process them with a <b>loop</b>.
</div>

```python
# STEP 1 — a small class with a decision method
class Learner:
    def __init__(self, name, cgpa):
        self.name = name
        self.cgpa = cgpa

    def grade(self):                 # logic: CGPA -> grade
        if   self.cgpa >= 3.5: return 'A'
        elif self.cgpa >= 3.0: return 'B'
        elif self.cgpa >= 2.0: return 'C'
        else:                  return 'F' 
```

```python
# STEP 2 — a list of objects, processed by a loop
batch = [
    Learner('Momina', 3.57),
    Learner('Amna',   3.10),
    Learner('Ahmad',  2.40),
    Learner('Ali',    1.80),
]

for L in batch:
    print(f"{L.name:8} CGPA={L.cgpa}  ->  Grade {L.grade()}")
```

```python
# STEP 3 — logic: average CGPA of only A/B students
good = [L.cgpa for L in batch if L.grade() in ('A', 'B')]
print("Top students :", len(good))
print("Average CGPA :", round(sum(good) / len(good), 2))
```

<div style="font-family:Segoe UI,system-ui,sans-serif;font-size:16px;color:#33475a;line-height:1.7">
Everything we learned about <b>classes</b> now builds a real <b>neural network</b>. It is just a bigger class:
<code style="background:#e6f0f6;color:#0a3d62;padding:2px 7px;border-radius:4px">__init__</code> sets up the weights,
<code style="background:#e6f0f6;color:#0a3d62;padding:2px 7px;border-radius:4px">train</code> teaches it,
<code style="background:#e6f0f6;color:#0a3d62;padding:2px 7px;border-radius:4px">query</code> asks it, and
<code style="background:#e6f0f6;color:#0a3d62;padding:2px 7px;border-radius:4px">backquery</code> shows what it imagined.
Run the import cell first.
</div>

```python
import numpy as np
import scipy.special

# compatibility: np.asfarray was removed in NumPy 2.0 — keep old notebooks working
if not hasattr(np, "asfarray"):
    np.asfarray = lambda a, dtype=float: np.asarray(a, dtype=dtype)
```

```python
class Neuralnetwork:
    def __init__(self, inodes, hnodes, onodes, learningrate):
        self.inputnodes=inodes
        self.hiddennodes=hnodes
        self.outputnodes=onodes
        self.lr=learningrate

        self.wih=np.random.normal(0.0, pow(self.hiddennodes, -0.5), (self.hiddennodes, self.inputnodes))
        self.who=np.random.normal(0.0, pow(self.outputnodes, -0.5), (self.outputnodes, self.hiddennodes))

        self.activation=lambda x: scipy.special.expit(x)
        self.inverse_activation_function = lambda x: scipy.special.logit(x)

        pass

    def train(self, model_input, labeled_output):

        inputs=np.array(model_input, ndmin=2).T
        lebeledouts=np.array(labeled_output, ndmin=2).T

        hidden_input=np.dot(self.wih, inputs)
        hidden_output=self.activation(hidden_input)

        final_input=np.dot(self.who, hidden_output)
        final_output= self.activation(final_input)

        output_error=lebeledouts-final_output
        hidden_error=np.dot(self.who.T,output_error)

        self.who+=self.lr*np.dot(output_error*final_output*(1-final_output), np.transpose(hidden_output))
        self.wih+=self.lr*np.dot(hidden_error*hidden_output*(1-hidden_output), np.transpose(inputs))
        pass

    def query(self,model_input):
        inputs=np.array(model_input, ndmin=2).T

        hidden_input=np.dot(self.wih, inputs)
        hidden_output=self.activation(hidden_input)

        final_input=np.dot(self.who, hidden_output)
        final_output= self.activation(final_input)
        return final_output

    #backwards query to have an idea what NN has learned
    def backquery(self, targets_list):
        # transpose the targets list to a vertical array
        final_outputs = np.array(targets_list, ndmin=2).T

        # calculate the signal into the final output layer
        final_inputs = self.inverse_activation_function(final_outputs)
        # calculate the signal out of the hidden layer
        hidden_outputs = np.dot(self.who.T, final_inputs)
        # scale them back to 0.01 to .99
        hidden_outputs -= np.min(hidden_outputs)
        hidden_outputs /= np.max(hidden_outputs)
        hidden_outputs *= 0.98
        hidden_outputs += 0.01

        # calculate the signal into the hidden layer
        hidden_inputs = self.inverse_activation_function(hidden_outputs)

        # calculate the signal out of the input layer
        inputs = np.dot(self.wih.T, hidden_inputs)
        # scale them back to 0.01 to .99
        inputs -= np.min(inputs)
        inputs /= np.max(inputs)
        inputs *= 0.98
        inputs += 0.01

        return inputs
```

  💡 Same idea as Dog and Student — a <b>class</b> with <code style="background:#e6f0f6;color:#0a3d62;padding:1px 6px;border-radius:4px">__init__</code> and methods. Just now the &quot;data&quot; are weight matrices, and the methods do the learning.

```python
# make one neural network object from the class
# (input nodes, hidden nodes, output nodes, learning rate)
NN = Neuralnetwork(3, 3, 3, 0.3)
print(NN)   # the object is ready
```
